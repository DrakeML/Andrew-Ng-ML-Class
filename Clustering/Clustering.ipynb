# Ward's Agglomerative Hierarchical Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
sns.set()
```

```python
from sklearn.datasets.samples_generator import make_blobs
```

```python
X, y = make_blobs(n_samples = 600, centers = 5, cluster_std = 0.6, random_state = 42)
```

```python
plt.scatter(X[:,0], X[:,1], s=10);
```

```python
from scipy.cluster.hierarchy import ward, dendrogram, linkage
```

```python
np.set_printoptions(precision = 4, suppress = True)
```

```python
distance = linkage(X, 'ward')
```

# Dendrogram

```python
plt.figure(figsize = (25,15))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel("Ward's distance")
dendrogram(distance, leaf_rotation = 90, leaf_font_size = 4);
```

# Truncating Dendrogram

```python
X, y = make_blobs(n_samples = 600, centers = 5, cluster_std = 0.6, random_state = 42)
distance = linkage(X, 'ward')
plt.figure(figsize = (25,15))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel("Ward's distance")
dendrogram(distance, leaf_rotation = 90, leaf_font_size = 4);
plt.axhline(25, c='k');
```

```python
plt.figure(figsize = (20,10))
plt.title('Hierarchical Clustering Dendrogram (truncated)')
plt.xlabel('Index')
plt.ylabel("Ward's distance")
dendrogram(distance, truncate_mode = 'lastp', p = 6,  leaf_rotation = 0, leaf_font_size = 12, show_contracted = True);
```

## Retrieve the Clusters

```python
from scipy.cluster.hierarchy import fcluster
```

```python
max_d = 150
clusters = fcluster(distance, max_d, criterion = 'distance')
clusters
```

```python
plt.figure(figsize=(10,8))
plt.scatter(X[:,0], X[:,1], c=clusters, cmap='prism');
```

# k-Means Clustering

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 5)
kmeans.fit(X)
```

```python
y_kmeans = kmeans.predict(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=y_kmeans, s=10, cmap='inferno')

centers = kmeans.cluster_centers_
plt.scatter(centers[:,0], centers[:,1], c='green', s=300, alpha = 0.7);
```

# Elbow Method

```python
kmeans.inertia_
```

```python
sse_ = []
for k in range(1,8):
    kmeans = KMeans(n_clusters = k).fit(X)
    sse_.append([k, kmeans.inertia_])
```

```python
plt.plot(pd.DataFrame(sse_)[0], pd.DataFrame(sse_)[1]);
```

# Silhouette Analysis

```python
from sklearn.metrics import silhouette_score
```

```python
sse_ = []
for k in range(2,8):
    kmeans = KMeans(n_clusters=k).fit(X)
    sse_.append([k, silhouette_score(X, kmeans.labels_)])
```

```python
plt.plot(pd.DataFrame(sse_)[0], pd.DataFrame(sse_)[1]);
```

# Mean Shift

```python
from sklearn.cluster import MeanShift, estimate_bandwidth
```

```python
from itertools import cycle
```

```python
bandwidth_X = estimate_bandwidth(X, quantile=0.1, n_samples=len(X))
```

```python
meanshift_model = MeanShift(bandwidth = bandwidth_X, bin_seeding=True, min_bin_freq=1,
                           n_jobs=1, seeds=None).fit(X)
```

```python
cluster_centers = meanshift_model.cluster_centers_
```

```python
print('\nCenters of clusters: \n', cluster_centers)
```

```python
labels = meanshift_model.labels_
num_clusters = len(np.unique(labels))
print('\nNumber of clusters in input data =', num_clusters)
```

```python
plt.figure(figsize=(10,8))
markers = '*vosx'
for i,marker in zip(range(num_clusters), markers):
    plt.scatter(X[labels==i, 0], X[labels==i, 1], marker = marker, color = 'orange')
    cluster_center = cluster_centers[i]
    plt.plot(cluster_center[0], cluster_center[1], marker='o', markerfacecolor='k',
            markeredgecolor='k', markersize=15)
plt.title('Clusters');
```
